### 2. _Reshaping data using Pandas Library._

- `Reshaping`: This is a process of changing the shape of the data. It can be done in many ways, some them are discussed below:
    - `Pivoting`: This is a process of transforming data from long format to wide format.
    - `Merging`: This is a process of combining data from two or more dataframes.
    - `Stack`: This is a process of stacking data from multiple rows into a single row.
    - `Unstack`: This is a process of unstacking data from a single row into multiple rows


#### **Melting_Dataframe**

Melting a DataFrame in Pandas refers to transforming it from a **wide format** to a **long format**, making it more suitable for analysis and visualization. It reshapes the data by converting column headers into row values, typically using the `pd.melt()` function. This is useful for unpivoting data, making it easier to group, filter, or plot.

```python
# Melting of the data frame using pandas library

df_3 = pd.melt(
    df_2, 
    id_vars = ['Name'], 
    value_vars = ['Age', 'Height_in_m', 'Weight_kg'], 
    var_name = 'Attributes', value_name = 'Values'
    )
df_3
```

### 3. _Pivot Tables using Pandas Library._

- `Pivot_tables`: A pivot table is a data summarization tool used to group, aggregate, and analyze large datasets efficiently. In Pandas, pivot tables allow you to restructure and summarize data based on different categories, similar to Excel pivot tables.

```python
pip install seaborn
```

```python
import seaborn as sns
import pandas as pd

# Load the Titanic dataset
titanic_df = sns.load_dataset("titanic")

# Creating a pivot table to analyze survival rates by class and gender
pivot_table = titanic_df.pivot_table(
    values="survived",  # The column to aggregate (0 = Died, 1 = Survived)
    index="class",      # Rows (Index) -> Passenger Class (First, Second, Third)
    columns="sex",      # Columns -> Gender (Male, Female)
    aggfunc="mean"      # Aggregation function -> Mean (Survival Rate)
)

# Display the pivot table
pivot_table
```

### 4. _Concatinating of data using Pandas Library._

- `Concatination`: Concatenating data means combining two or more data structures into one. Pandas library provides a function
`concat()` to concatenate data. It can be used to concatenate DataFrames, Series, or even other data structures.

```python
# Here we simply use the concat commant to join the two dataframes into a single dataframe.
# Creating first DataFrame
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie']
})

# Creating second DataFrame
df2 = pd.DataFrame({
    'ID': [4, 5, 6],
    'Name': ['David', 'Eva', 'Frank']
})

# Concatenating DataFrames Vertically (Row-wise)
vertical_concat = pd.concat([df1, df2], axis=0, ignore_index=True)
vertical_concat
```

```python
# Concatenating DataFrames Horizontally (Column-wise)
horizontal_concat = pd.concat([df1, df2], axis=1)
horizontal_concat
```

9. `Var() function`: This function returns the variance of the data. It is used to calculate the variance of the data. variance is generally used to measure the spread of the data. It is calculated as the average of the squared differences from the mean. The formula for variance is: variance = Σ(xi – μ)2 / (n – 1) where xi is the individual data point, μ is the mean, and n is the number of data points. The variance is a measure of the spread of the data and is used in many statistical tests. The variance is always non-negative and is zero if and only if all th data points are equal.


```python
df_2['Weight_kg'].var()
```

10. `Std() function`: This function returns the standard deviation of the data. It is used to calculate the standard deviation of th data. Standard deviation is a measure of the amount of variation or dispersion of a set of values. A low standard deviation indicates that the values tend to be close to the mean (also called the expected value) of th data set, while a high standard deviation indicates that the values are spread out over a wider range.  The formula for standard deviation is: std = sqrt(sum((x_i - mean)^2) / (n - 1)) where x_i is each value in the data set, mean is th mean of the data set, n is the number of values in the data set, and sqrt is the square root function.

```python
df_2['Weight_kg'].std()
```

### 15. _Combine Datasets using Python Library._

- Combine the data frame using the merge function. This will combine the two data frames based on the common column. In this case,we are merging the data frames df1 and df2 dataset.

```python
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie']
})

df2 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['David', 'Eva', 'Frank']
})
```

```python
# merge the two data frames using the merge function, along left axis.
pd.merge(df1, df2, on='ID', how='left')
```

```python
# merge the two data frames using the merge function, along right axis.
pd.merge(df1, df2, on='ID', how='right')
```

```python
# merge the two data frames using the merge function, along inner axis.
pd.merge(df1, df2, on='ID', how='inner')
```

```python
# merge the two data frames using the merge function, along outer axis.
pd.merge(df1, df2, on='ID', how='outer')
```

### 17. _Plotting using Python Library._

- `Plotting Data`: Plotting data is a crucial step in data analysis. It helps in visualizing the data and understanding the relationships between different variables. Python has several libraries that can be used for plotting data, including `matplotlib` and `seaborn`. 

```python
# Creating Histogram using pandas library.

import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = np.random.randn(1000)  # 1000 random values from a normal distribution

# Create histogram
plt.hist(data, bins=30, color='green', edgecolor='black', alpha=0.7)

# Labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')

# Show plot
plt.show()
```

```python
# Creating Bar plot using pandas library.
import matplotlib.pyplot as plt

# Sample data
categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 24, 36, 18, 30]

# Create bar plot
plt.bar(categories, values, color='green', alpha=0.7)

# Labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot Example')

# Show plot
plt.show()
```

```python
# Creating scatter plot using pandas library.
# Generate sample data
x = np.random.rand(100) * 10  # 100 random x values between 0 and 10
y = np.random.rand(100) * 10  # 100 random y values between 0 and 10

# Create scatter plot
plt.scatter(x, y, color='green', alpha=0.7, edgecolors='black')

# Labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')

# Show plot
plt.show()
```
